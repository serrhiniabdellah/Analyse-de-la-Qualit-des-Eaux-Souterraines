```python
#question1

import pandas as pd
import math as mt
from functools import reduce
import numpy as np

# Load the WaterQuality dataset into a pandas DataFrame.
df=pd.read_excel('WaterQuality.xlsx')
# Load the Paramètres1 dataset into a pandas DataFrame.
dfp1=pd.read_excel('Paramètres.xlsx')
dfp1.index=['Si','wi']
dfp1
```

                              Moyenne arithmetique pondérée (WAWQI)

```python
#question2

# Define a function called Wi to calculate the weight of each parameter.
def Wi():
    list1=[]
    for i in dfp1.iloc[1,:]:
        S=0
        for j in dfp1.iloc[1,0:]:
            S=S+j
        Wi=i/S
        list1.append(Wi)
    dfp1.loc["Wi",:]=list1

    # Call the Wi function to calculate the weights.
Wi()
dfp1
```

```python
#question2

# Call the Wi function again.
Wi()

# Define a function called WAWQI to calculate the Weighted Arithmetic Water Quality Index (WAWQI).
def WAWQI():
    list1=[]
    for i in range(0,18):
        list2=[]
        for key, value in df.items():
            if key=="pH":
              v=value[i]
              Qi=(v-8.5)/(6.5-8.5)
            elif key=="FC" or key=="Year" or key=="POINTS": continue
            else:
              v=value[i]
              Qi=(v/dfp1.loc["Si",key])*100
            SII=Qi*dfp1.loc["Wi",key]
            list2.append(SII)
        list1.append(sum(list2))
    df.loc[:,"WAWQI"]=list1
# Call the WAWQI function to calculate the WAWQI.
WAWQI()
df
```

```python
#question3

# Define a function called interpretation to interpret the WAWQI values.
def interpretation():
    list1=[]
    for i in df.loc[:,"WAWQI"]:
       if i<50:
           q="Excellent Water"
       elif i>=50 and i<=100:
           q="Good Water"
       elif i>=101 and i<=200:
           q="Poor Water"
       elif i>=201 and i<=300:
           q="Very Poor water"
       else:
           q="Unsuitable for drinking"
       list1.append(q)
    df.loc[:,"Quality"]=list1

# Call the interpretation function to interpret the WAWQI values.
interpretation()
df
```

                            Moyenne géométrique pondérée (WGWQI)

```python
#question 5

# Call the Wi function to calculate the weights.
Wi()

def WGWQI():
    list1=[]
    for i in range(0,18):
        list2=[]
        for key, value in df1.items():
            if key=="pH":
                v=value[i]
                Qi=(v-8.5)/(6.5-8.5)
            elif key=="FC" or key=="Year" or key=="POINTS":
                continue
            else:
                v=value[i]
                Qi=(v/dfp1.loc["Si",key])*100
            qiwi=mt.pow(Qi,dfp1.loc["Wi",key])
            list2.append(qiwi)
        list1.append(reduce(lambda x,y:x*y,list2))
    df1.loc[:,"WGWQI"]=list1

# Call the WGWQI function to calculate the WGWQI.
WGWQI()
df1
```

```python
#question6

# Define a function called interpretation to interpret the WGWQI values.
def interpretation():
    list1=[]
    for i in df1.loc[:,"WGWQI"]:
       if i>=90 and i<=100:
           q="Excellent"
       elif i>=70 and i<=89:
           q="Good"
       elif i>=50 and i<=69:
           q="Medium"
       elif i>=25 and i<=49:
           q="Bad"
       else:
           q="Very Bad"
       list1.append(q)
    df1.loc[:,"Quality"]=list1
# Call the interpretation function to interpret the WGWQI values.

interpretation()
df1
```

                       L'indice d'Oregon de Qualité de l'eau (OWQI)

```python
#question8

# Load the Paramètres1 dataset into a pandas DataFrame.
dfp1=pd.read_excel('Paramètres.xlsx')

# Set the index of the dfp1 DataFrame to 'Si' and 'wi'.
dfp1.index=['Si','wi']

# Add a new column called 'FC' to the dfp1 DataFrame.
list1=[594,5]
dfp1.loc[:,"FC"]=list1

# Remove the columns 'DCO', 'CONDUCTIVITE', 'MES', 'TURBIDITE', and 'NITRITE' from the dfp1 DataFrame.
list2=["DCO","CONDUCTIVITE","MES","TURBIDITE","NITRITE"]
for i in list2:
    dfp1.pop(i)


# Call the Wi function to calculate the weights.
Wi()
dfp1
```

```python
#question8
# Define a function called OWQI to calculate the Oregon Water Quality Index (OWQI).
def func(a,b):
    try:
        return a/b
    except ZeroDivisionError as e:
        return np.inf
```

```python
#question8
def OWQI():
        list1=[]
        for i in range(0,18):
            list2=[]
            for key, value in df2.items():
                if key=="pH":
                    v=value[i]
                    Qi=(v-8.5)/(6.5-8.5)
                elif key=="DCO" or key=="Year" or key=="POINTS" or key=="CONDUCTIVITE" or key=="MES" or key=="TURBIDITE" or key=="NITRITE": continue
                else:
                    v=value[i]
                    Qi=(v/dfp1.loc["Si",key])*100
                SII=Qi*dfp1.loc["Wi",key]
                list2.append(func(1,mt.pow(SII,2)))
            list1.append(mt.sqrt(8/sum(list2)))
        df2.loc[:,"OWQI"]=list1

# Call the OWQI function to calculate the OWQI.
OWQI()
df2
```

```python
#question9
# Define a function called interpretation to interpret the OWQI values.
def interpretation():
    list1=[]
    for i in df2.loc[:,"OWQI"]:
       if i>=90 and i<=100:
           q="Excellent"
       elif i>=85 and i<=89:
           q="Good"
       elif i>=80 and i<=84:
           q="Fair"
       elif i>=60 and i<=79:
           q="Poor"
       else:
           q="Very Poor"
       list1.append(q)
    df2.loc[:,"Quality"]=list1

# Call the interpretation function to interpret the OWQI values.
interpretation()
df2
```
